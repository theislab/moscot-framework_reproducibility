```r
library("stringr")
adapt_obs_names <- function(x){
    tmp <- str_split(x, "-")[[1]]
    if(tmp[3] == "0"){
        prefix <- "E14-5"
        }
    else{
        prefix <- "E15-5"
        }
    return(paste(prefix, "_", tmp[1], "-", tmp[2], sep = ""))
    }
```

```r
adata$obs$obs_name_adapted <- sapply(adata$obs$index_name, adapt_obs_names) 
```

```r
seurat <- seurat[,adata$obs$obs_name_adapted]
```

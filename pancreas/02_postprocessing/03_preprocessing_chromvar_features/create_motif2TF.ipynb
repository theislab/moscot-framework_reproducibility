```python
import pandas as pd
import os
```

```python
files_list = os.listdir("/Users/dominik.klein/Downloads/Mus_musculus_2023_01_03_10_03_am/pwms_all_motifs")
```

```python
motifs = [files_list[i][:-4] for i in range(len(files_list))]
```

```python
df = pd.read_csv("/Users/dominik.klein/Downloads/Mus_musculus_2023_01_03_10_03_am/TF_Information_all_motifs_plus.txt", delimiter = "\t")
```

```python
df_present = df[df["Motif_ID"].isin(motifs)]
```

```python
motif2TF = df_present.groupby("Motif_ID").TF_Name.unique()
```

```python
filter_list = pd.read_csv("filtering_list.csv")
```

```python
df_present
```

```python
motif2TF_filtered = motif2TF[~filter_list["x"].values]
df_present["direct_TF"] = df_present.apply(lambda x: None if x["TF_Status"] == "I" else x["TF_Name"], axis=1)
df_motif2TF = pd.DataFrame(motif2TF_filtered)
df_present.set_index("Motif_ID", inplace=True)
```

```python
df_present_red = df_present[df_present["TF_Status"] == "D"]
```

```python
df_merged = pd.merge(df_motif2TF, df_present_red[["direct_TF"]], how="left", left_index=True, right_index=True)
```

```python
df_merged.to_csv("motif2TF.csv")
```

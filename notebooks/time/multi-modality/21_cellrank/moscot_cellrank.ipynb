```python
import scanpy as sc
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import moscot
from moscot.problems.time import TemporalProblem
import moscot.plotting as mpl
import pandas as pd
import os

sc.set_figure_params(scanpy=True, dpi=80, dpi_save=200)

import mplscience
from cellrank.kernels import RealTimeKernel
mplscience.available_styles()
mplscience.set_style(reset_current=True)
plt.rcParams['legend.scatterpoints'] = 1 
```

```python
tp0 = TemporalProblem.load("/lustre/groups/ml01/workspace/moscot_paper/pancreas_revision/plots/OT_encodrine_analysis/TemporalProblem.pkl")
```

```python
tmk = RealTimeKernel.from_moscot(tp0)
```

```python
sc.pp.neighbors(tp0.adata, use_rep="X_MultiVI")
tmk.compute_transition_matrix(self_transitions="all", conn_weight=0.001, threshold="auto")
```

```python
import cellrank
g = cellrank.estimators.GPCCA(tmk)
```

```python
g.fit(cluster_key="cell_type", n_states=25)
```

```python
g.set_terminal_states(states=["Alpha_1", "Beta_1", "Epsilon", "Delta_2"])
```

```python
g.compute_fate_probabilities()
```

```python
g.plot_fate_probabilities(same_plot=False)
```

```python
cellrank.pl.aggregate_fate_probabilities(
        adata=tp0.adata,
        mode='heatmap',
        lineages=["Alpha_1", "Beta_1", "Epsilon", "Delta_2"],
        cluster_key="cell_type",
        clusters=None,
        figsize=(20, 6),
        dpi=300,
    )
```

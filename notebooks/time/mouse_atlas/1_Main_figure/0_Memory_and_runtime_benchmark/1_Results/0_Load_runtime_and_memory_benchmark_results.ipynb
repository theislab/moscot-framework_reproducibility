```python
import pandas as pd
```

```python
cols_kept=['Method', 'k', 'Max_memory', 'Sinkhorn_time', 'Evaluation_time', 'Apoptosis_rate',
          'Accuracy_Curated', 'Accuracy_Germ', 'Rank', 'Gamma', 'tau1']
```

```python
# From this run I only need WOT (I don't use the moscot-runs since I used a subideal batch_size)
df0=pd.read_csv('wandb_WOT.csv')
df0=df0[df0['Method']=='WOT'][cols_kept]
df0=df0[df0['k'].isin([25,50,75,100])]
df0.index=range(len(df0))
df0
```

```python
# From this run I only take the GPU runs, since the CPU runs are the GPU-node CPUs and they create a OOM error
# Low-rank is rank 2000 here
df1=pd.read_csv('wandb_GPU.csv')
df1=df1[df1['Method'].isin(['moscot_LR_GPU', 'moscot_CPU'])][cols_kept]
# I mixed up CPU and GPU naming, i.e. moscot_CPU is actually moscot_GPU
df1['Method']=[a if a=='moscot_LR_GPU' else 'moscot_GPU' for a in df1['Method']]
df1.index=range(len(df1))
df1.sort_values('k')
```

```python
# And these are the CPU runs
df2=pd.read_csv('wandb_CPU.csv')
df2=df2[cols_kept]
df2.index=range(len(df2))
df2.sort_values('k')
```

```python
df=pd.concat([df0, df1, df2], ignore_index=True)
df['tau2']=0.9995
df['eps']=[0.0001 if 'LR' in a else 0.005 for a in df['Method']]
df.sort_values('k')
```

```python
df.to_pickle('benchmark_results.pkl')
```

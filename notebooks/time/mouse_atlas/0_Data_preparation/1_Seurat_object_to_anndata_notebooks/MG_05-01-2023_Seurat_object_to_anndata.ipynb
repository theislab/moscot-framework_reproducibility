# RDS file to AnnData

Runs SeuratDisk/Data to convert the downloaded RDS-files into AnnData objects

```python
library(Seurat)
library(SeuratData)
library(SeuratDisk)
```

```python
sessionInfo()
```

```python
file_path='/home/mgander/moscot-framework_reproducibility/data/mouse_temporal/RDS'

work_path = "/home/mgander/moscot-framework_reproducibility/data/mouse_temporal/"

time_point = paste0("E", c(3.5, 4.5, 5.25, 5.5, 6.25, seq(6.5, 8.25, 0.25), "8.5a", "8.5b", seq(9.5, 13.5, 1)))
```

```python
for (i in 1:20) {
    time_1=time_point[i]

    if (time_1=='E7')
        {time_1='E7.0'}

    Exp = readRDS(paste0(file_path, "/seurat_object_", time_1, ".rds"))
    meta=Exp@meta.data
    
    if (time_1=='E8')
        {time_1='E8.0'}

    write.csv(x=meta, file=paste0(work_path, "/metadata/meta_", time_1, ".csv"))
    SaveH5Seurat(Exp, filename = paste0(work_path, "anndatas/Single_anndatas/adata_", time_1, ".h5Seurat"))
    Convert(paste0(work_path, "anndatas/Single_anndatas/adata_", time_1, ".h5Seurat"), dest = "h5ad")
}
```

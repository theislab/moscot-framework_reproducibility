```python
genes=['Gata4', 'Amot', 'Hnf1b', '...']
mc=[0.73, 0.78, 0.78, '...']
tc=[0.51, 0.65, 0.63, '...']
```

```python
k=4
dfm=pd.DataFrame({'Method':['moscot']})
for i in range(k):
    dfm[f'{genes[i]}']=mc[i]
dft=pd.DataFrame({'Method':['TOME']})
for i in range(k):
    dft[f'{genes[i]}']=tc[i]

df=pd.concat([dfm,dft], ignore_index=True)
```

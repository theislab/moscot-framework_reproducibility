```python
Path="/home/mgander/Unrelated/moscotTime_Reproducibility/Data"
ts=['E3.5', 'E4.5', 'E5.25', 'E5.5', 'E6.25', 'E6.5', 'E6.75', 'E7.0', 'E7.25', 'E7.5', 'E7.75', 'E8.0', 'E8.25', 'E8.5a', 'E8.5b', 'E9.5', 'E10.5', 'E11.5', 'E12.5', 'E13.5']
```

# Moscot maps to growth rates

```python
pec_perc=[]

for i in range(9,13):
    ts0=ts[i]
    ts1=ts[i+1]
    adata=sc.read(f"{Path}/anndatas/Comb_anndatas/adata_{ts0}_{ts1}.h5ad")
    t1=max(set(adata.obs['day']))
    A1=adata[adata.obs['day']==t1].copy()
    pec_perc.append(len(A1[A1.obs['cell_type']=='Primitive erythroid cells'])/len(A1))
```

```python
df=pd.DataFrame({'Time point':['E7.75', 'E8.0', 'E8.25', 'E8.5'], 'PEC_Perc':pec_perc})
```
